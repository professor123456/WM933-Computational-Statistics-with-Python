```python
# use DNN to predict the result, use dropout and RMSprop to fit the overfitting problem
# use L2 to improve model accuracy
import tensorflow as tf
from tensorflow import keras
from keras import regularizers
from tensorflow.keras import layers
def applications_total_model():
    class myCallback(tf.keras.callbacks.Callback):
        def on_train_begin(self, logs={}):
            self.losses = []
        def on_batch_end(self, batch, logs={}):
            self.losses.append(logs.get('loss'))
    callbacks = myCallback()
# record loss history    
#source:https://keras.io/zh/callbacks/
    model_DNN = keras.Sequential([
      layers.Dense(500, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01), input_shape =(111,)),
      layers.Dropout(0.4),
      layers.Dense(500, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01)),
      layers.Dropout(0.4),
      layers.Dense(500, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01)),
      layers.Dropout(0.4),
      layers.Dense(500, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01)),
      layers.Dropout(0.4),
      layers.Dense(250, activation = tf.nn.tanh),
      layers.Dense(1)
  ])
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    model_DNN.compile(loss = "mean_squared_error" ,optimizer = optimizer, metrics =['mae', 'mse'])
    model_DNN.fit(x_train, y_train, batch_size=128, epochs=500, verbose=1, callbacks=[callbacks])
    return model_DNN
```

```python
model_DNN = applications_total_model()
model_DNN.summary()
```
